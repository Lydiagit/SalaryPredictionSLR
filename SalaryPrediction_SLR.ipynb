# 5. Create testing and training dataset

```python
# Lets name the data
x =df[['YearsExperience']]
```

```python
y = df[['Salary']]
```

```python
#import sklearn to use in splitting the data

from sklearn.model_selection import train_test_split 
#train_test_split tab shift for a detailed info on how to use this class - it shuffles and split the data
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)                 
```

```python
y_test
```

```python
y_train.shape
```

# 6. Train the Model

```python
#import linearregresion
# to train using linear regression
from sklearn.linear_model import LinearRegression 
```

```python
regressor = LinearRegression(fit_intercept = True)
regressor.fit(x_train, y_train)
```

```python
# the model is now trained, what we need is to get/ print   m and b

print ('Linear Model Coefficient (m)', regressor.coef_)
print ('Linear Model Coefficient (b)', regressor.intercept_)
```

# 7. Test the Model

```python
regressor
```

```python
#To predict y
y_predict = regressor.predict(x_test)
```

```python
y_predict
```

```python
# Lets visualize test data vs the predicted data using our model
plt.scatter(x_test, y_test, color = 'gray')
plt.plot(x_test, regressor.predict(x_test), color = 'red')
plt.xlabel('years')
plt.ylabel('salary')
plt.title('test data')
```

 using the least squared method our model perfectly predicts the y values of our test datasets
